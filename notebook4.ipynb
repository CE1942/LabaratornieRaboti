```python
import pandas as pd

# Создание Series
data = {'A': 12, 'B': 24, 'C': 36, 'D': 48, 'E': 60}
series = pd.Series(data)

# Получение значения по индексу 'C'
value_by_loc = series.loc['C']

# Получение третьего элемента по позиции (нумерация начинается с 0)
third_element = series.iloc[2]

# Вывод результатов
print(f"Значение по индексу 'C': {value_by_loc}")
print(f"Третий элемент по позиции: {third_element}")
```

Задание 5

```python
import pandas as pd

# Создаем Series из списка животных
animals = pd.Series(['cat', 'dog', 'rabbit', 'parrot', 'fish'])

# Проверяем исходный тип данных
original_dtype = animals.dtypes

# Преобразуем тип данных в категорию
animals_category = animals.astype('category')

# Проверяем новый тип данных
new_dtype = animals_category.dtypes

# Выводим исходный и преобразованный типы данных
print(f"Исходный тип данных: {original_dtype}")
print(f"Преобразованный тип данных: {new_dtype}")
```

```python
import pandas as pd
import numpy as np

# Создаем первый Series
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# Создаем второй Series
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

# Выполняем операцию сложения
result = s1 + s2

# Заменяем NaN на 0
result_filled = result.fillna(0)

# Выводим результат
print(result_filled)
```

Задание 10

```python
import numpy as np
import pandas as pd

# Генерация 20 случайных чисел от 50 до 150
random_data = np.random.randint(50, 151, size=20)

# Создание Series
s = pd.Series(random_data)

# Расчет статистик
sum_value = s.sum()
mean_value = s.mean()
min_value = s.min()
max_value = s.max()
std_deviation = s.std()

# Вывод результатов
print(f"Сумма: {sum_value:.2f}")
print(f"Среднее: {mean_value:.2f}")
print(f"Минимальное значение: {min_value}")
print(f"Максимальное значение: {max_value}")
print(f"Стандартное отклонение: {std_deviation:.2f}")
```

Задание 12

```python
import pandas as pd

# Создаем Series
s = pd.Series([10, 20, 30, 40, 50, 60], index=['A', 'B', 'A', 'C', 'D', 'B'])

# Проверяем уникальность индексов
is_unique = s.index.is_unique

if not is_unique:
    # Группируем по индексам и суммируем значения
    grouped_sum = s.groupby(level=0).sum()
else:
    print("Индексы уникальны.")

# Выводим итоговую группу
grouped_sum
```

```python
import pandas as pd

# Данные для записи в CSV
data = {
    'Дата': ['2024-03-01', '2024-03-02', '2024-03-03', '2024-03-04', '2024-03-05'],
    'Цена': [100, 110, 105, 120, 115]
}

# Сохраняем данные в CSV-файл
df = pd.DataFrame(data)
df.to_csv('data.csv', index=False)

# Читаем CSV-файл
df = pd.read_csv('data.csv')

# Создаем Series, используя "Дата" в качестве индекса
prices = df.set_index('Дата')['Цена']

# Выводим полученный Series
print(prices)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Создаем индексы дат с 1 по 30 марта 2024 года
dates = pd.date_range(start="2024-03-01", end="2024-03-30")

# Создаем Series с случайными значениями от 50 до 150
values = np.random.randint(50, 151, len(dates))
series = pd.Series(values, index=dates)

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(series.index, series.values, marker='o', linestyle='-', color='blue')

# Настраиваем оформление графика
plt.title('График случайных значений за март 2024 года')
plt.xlabel('Дата')
plt.ylabel('Значения')
plt.grid(True)

# Выводим график
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Создаем Series с датами и случайными значениями
dates = pd.date_range(start='2024-04-01', periods=15, freq='D')
values = np.random.randint(500, 1001, size=15)
series = pd.Series(values, index=dates)

# Рассчитываем среднее значение
mean_value = series.mean()

# Рисуем график
plt.figure(figsize=(10, 6))
plt.plot(series.index, series.values, label='Данные', marker='o', linestyle='-')
plt.axhline(y=mean_value, color='r', linestyle='--', label=f'Среднее ({mean_value:.2f})')

# Оформляем график
plt.title('График значений с пунктом среднего значения')
plt.xlabel('Дата')
plt.ylabel('Значения')
plt.legend()
plt.grid(True)

# Показываем график
plt.show()
```
